```python
# Exercise 11
import pandas as pd
from scipy import stats

# Calculate Z-scores for each data point
z_scores = stats.zscore(x)

# Define threshold for outlier detection (e.g., Z-score > 3 or < -3)
threshold = 3

# Filter rows where absolute Z-score exceeds the threshold
x_no_outliers = x[(abs(z_scores) < threshold).all(axis=1)]
```

```python
# solution with only pandas
import pandas as pd

# Assuming df is your DataFrame

# Calculate the first quartile (Q1) and third quartile (Q3) for each column
Q1 = Train.quantile(0.25)
Q3 = Train.quantile(0.75)

# Calculate the interquartile range (IQR) for each column
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df_no_outliers = Train[~((Train < lower_bound) | (Train > upper_bound)).any(axis=1)]
```
